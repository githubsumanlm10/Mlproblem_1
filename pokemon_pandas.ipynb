```python
import re
```

```python
# CHECKING STAING CONTAINING IN COLUMNS

df.loc[df['Type 1'].str.contains('Grass|Fire', regex=True)]
```

```python
# WE CAN USE FLAGS TO IGNORE CASE SENSITIVITY ( CAPITAL_LETTER = small letter)

df.dropna(subset=['Type 2'], inplace=True)

df.loc[df['Type 2'].str.contains('water|poison' , regex=True, flags=re.I)]
```

```python
# CHECKING THE COLUMNS IF ANY NAME STARTS WITH SPECIFIC STRING NAME LIKE PI, SE, ETC 
#  ^x THIS SIGN MEANS THE WORD WIL BE STARTING FROM x 
#  * MEANS ZERO OR MORE
#  ^pi[a-z]*  MEANS STARTING FROM PI THEN NEXT LETTERS CAN BE BETWEEN A TO Z AND ZERO OR MORE WORD CAN COME AFTER PI
# IF WE DON'T GIVE ^ THEN THAT WILL FIND 'PI' ANYWHERE IN THE NAME & IF WE DON'T GIVE '*' THEN IT WILL ONLY GET PI NOT ANYTHING COMES AFTER THAT


df.loc[df['Name'].str.contains('^pi[a-z]*' , flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('pi[a-z]*' , flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]' , flags=re.I, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi' , flags=re.I, regex=True)]
```
