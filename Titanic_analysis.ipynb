```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
def compute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
    
    
df['Age']=df[['Age', 'Pclass']].apply(compute_age, axis=1)
```

```python
sex=pd.get_dummies(df['Sex'], drop_first=True)
embark=pd.get_dummies(df['Embarked'], drop_first=True)
```

```python
df.drop(columns=['Sex', 'Embarked', 'Name', 'PassengerId', 'Ticket', 'Fare', 'Cabin'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df =pd.concat([df, sex, embark], axis=1)
```

```python
df
```

```python
x= df.drop('Survived', axis=1)
x
```

```python
y=df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, random_state=1, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(x_train, y_train)
```

```python
predictons=lr.predict(x_test)
```

```python
predictons
```

```python
from sklearn.metrics import confusion_matrix

a=confusion_matrix(y_test, predictons.round())
a
```
